```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

# Sample dataset with missing values
df = pd.DataFrame({
    "age": [25, np.nan, 47, 51, 38],
    "income": [50000, 60000, np.nan, 120000, 75000],
    "city": ["New York", "Los Angeles", "Chicago", np.nan, "Chicago"]
})


#hadling missing values
x = df['age'].mean()
y = df['income'].mean()
df['age'].fillna(x,inplace=True)
df['income'].fillna(y,inplace=True)
df['city'].fillna('Chicago',inplace=True)
df
```

```python
#hadling missing values with transformer
# num = SimpleImputer(strategy='mean')
# cat = SimpleImputer(strategy='most_frequent')
# df[['age','income']]=num.fit_transform(df[['age','income']])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), ["age", "income"]),
        ("cat", OneHotEncoder(), ["city"]),
    ]
)

x = preprocessor.fit_transform(df)
nf = ['age','income']
cf = preprocessor.named_transformers_['cat'].get_feature_names_out(['city'])
af = list(nf) + list(cf)

xdf = pd.DataFrame(x,columns=af)
xdf[cf] = xdf[cf].astype(int)
xdf
```
