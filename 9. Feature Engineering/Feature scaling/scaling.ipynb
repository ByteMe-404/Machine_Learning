```python
import seaborn as sn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
data = {
    "Height_cm": [150, 160, 170, 180, 190],
    "Weight_kg": [50, 65, 80, 95, 110],
    "Exam_Score": [55, 60, 65, 85, 95]
}

df = pd.DataFrame(data)
df
```

```python
scaler = StandardScaler() #can change scale here
df_standard = pd.DataFrame(scaler.fit_transform(df[["Height_cm"]]))
df_standard.describe()
df_standard
```
