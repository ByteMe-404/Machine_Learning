```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Ordinal encoding is a method of converting categorical variables with a **natural order** into numerical values while preserving that order.

# When to Use
# - Categories have a natural order :
#   - Education: High School < Bachelor < Master < PhD
#   - Ratings: Poor < Average < Good < Excellent
# - Algorithms that can work with integers: linear models, tree-based models, etc.

# ---
df = pd.DataFrame({
    'Education': ['High School', 'Bachelor', 'Master', 'PhD', 'Bachelor', 'Master', 'High School'],
    'Experience': ['Junior', 'Mid', 'Senior', 'Senior', 'Mid', 'Mid', 'Junior'],
    'Age':[17,27,33,22,25,26,20]
})
df
```

```python
edu_ord = ['High School','Bachelor','Master','PhD']
exp_ord = ['Junior','Mid','Senior']
```

```python
encoder = OrdinalEncoder(categories=[edu_ord,exp_ord])
df[['Education','Experience']] = encoder.fit_transform(df[['Education','Experience']])
df
```
