```python
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
import kagglehub
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
df['Age'].skew()
```

# Outlier detection


```python
q1 = df['Age'].quantile(0.25)
q2 = df['Age'].quantile(0.75)
iqr = q2 -q1
lower_bound = q1 - 1.5*iqr
upper_bound = q1 + 1.5*iqr

print(q1)
print(q2)
print(iqr)


outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]
display(outliers)
```

# Trimming outliers


```python
df_trimmed_iqr = df[(df["Age"] >= lower_bound) & (df["Age"] <= upper_bound)].copy()

print("Dataset after trimming:")
df_trimmed_iqr.head()
```

# Capping

```python
# Replace outliers with nearest bound
df["value_capped_iqr"] = df["Age"].clip(lower_bound, upper_bound)

print("Dataset after capping:")
df.head()
```
